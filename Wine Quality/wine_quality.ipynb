```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn import svm

import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
from collections import Counter
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score,classification_report
print("All dependencies are imported.")
```

```python
data = pd.read_csv('winequality-red.csv')
df = pd.DataFrame(data)
```

```python
#Output variable:
target_variables=df.iloc[:,-1]
display(target_variables)
```

```python
target_variables.unique()
```

```python
print(" ----------Input Variables------------")
x=df.iloc[:,:-1]
display(x)
```

```python
x.info()
```

```python
y = target_variables
```

```python
print("Model Parameters |  Accuracy Score  | Precision  | Recall  | F1 Score  | Classification Report")
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,stratify=y, random_state=23)
print("Parameters:")
display(LogisticRegression())
lr=LogisticRegression().fit(x_train,y_train)
y_pred=lr.predict(x_test)
print('Accuracy Score:')
print(round(accuracy_score(y_test,y_pred)*100,2))
print('\nPrecision:')
print(round(precision_score(y_test,y_pred, average='weighted')*100,2))
print('\nRecall:')
print(round(recall_score(y_test,y_pred, average='weighted')*100,2))
print('\nF1 Score:')
print(round(f1_score(y_test,y_pred, average='weighted')*100,2))
```
